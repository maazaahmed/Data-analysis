# Loading Data from PDF to a Dataframe


```python
import camelot

camelot_df = (camelot
              .read_pdf("crimes_reported_by_type_and_provinces.pdf",
                        flavor="stream",
                        suppress_stdout=True,
                        pages="all")
              [0].df)
```

```python
camelot_df
```

# Making First Row as Fields


```python
new_header = camelot_df.iloc[0] #grab the first row for the header
camelot_df = camelot_df[1:] #take the data less the header row
camelot_df.columns = new_header #set the header row as the df header
```

```python
camelot_df
```

# Removing Irrelevant Rows from the Dataframe

```python
camelot_df =  camelot_df.drop([1,1])
camelot_df = camelot_df.drop([12,12])
camelot_df = camelot_df.drop([23,23])
```

```python
camelot_df
```

# Adding Year Field at the of the Dataframe

```python
camelot_df.loc[camelot_df.index[0:10], 'Year'] = "2017"
camelot_df.loc[camelot_df.index[10:20], 'Year'] = "2018"
camelot_df.loc[camelot_df.index[20:30], 'Year'] = "2019"
```

```python
camelot_df
```

# Dump Data into the Postgresql Database

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:51214@localhost:5432/testdb')
camelot_df.to_sql('crimeRecord', engine)
```

```python
my_dict = {"Punjab":[],"Sindh":[],"KP":[], "Balochistan":[], "Islamabad":[], "Railways":[], "G.B":[], "AJK":[]};
```

```python
try:
    connection = psycopg2.connect(user="postgres",
                                  password="51214",
                                  host="127.0.0.1",
                                  port="5432",
                                  database="testdb")
    cursor = connection.cursor()
    postgreSQL_select_Query = '''SELECT * from "crimeRecord" WHERE "Offence" = 'TOTAL RECORDED CRIME';'''

    cursor.execute(postgreSQL_select_Query)
    print("Selecting rows from crime record table using cursor.fetchall")
    records = cursor.fetchall()

    print("Print each row and it's columns values")
    for row in records:
        my_dict["Punjab"].append(int(row[2]))
        my_dict["Sindh"].append(int(row[3]))
        my_dict["KP"].append(int(row[4]))
        my_dict["Balochistan"].append(int(row[5]))
        my_dict["Islamabad"].append(int(row[6]))
        my_dict["Railways"].append(int(row[7]))
        my_dict["G.B"].append(int(row[8]))
        my_dict["AJK"].append(int(row[9])
        

except (Exception, psycopg2.Error) as error:
    print("Error while fetching data from PostgreSQL", error)


cursor.close() 

        
```

```python
total_dict = {"Punjab": 0,"Sindh": 0,"KP": 0, "Balochistan": 0, "Islamabad": 0, "Railways": 0, "G.B": 0, "AJK": 0};
```

```python
total_dict["Punjab"] = sum(my_dict["Punjab"])
total_dict["Sindh"] = sum(my_dict["Sindh"])
total_dict["KP"] = sum(my_dict["KP"])
total_dict["Balochistan"] = sum(my_dict["Balochistan"])
total_dict["Islamabad"] = sum(my_dict["Islamabad"])
total_dict["Railways"] = sum(my_dict["Railways"])
total_dict["G.B"] = sum(my_dict["G.B"])
total_dict["AJK"] = sum(my_dict["AJK"])
```

```python
import operator
print(max(total_dict.items(), key=operator.itemgetter(1))[0], "have the highest crime rate")
```

```python
year_dict =  {"2017": 0,"2018": 0,"2019": 0};
```

```python
try:
    connection = psycopg2.connect(user="postgres",
                                  password="51214",
                                  host="127.0.0.1",
                                  port="5432",
                                  database="testdb")
    cursor = connection.cursor()
    postgreSQL_select_Query = '''SELECT "Pakistan" from "crimeRecord" WHERE "Offence" = 'TOTAL RECORDED CRIME';'''

    cursor.execute(postgreSQL_select_Query)
    records = cursor.fetchall()

    print("Print each row and it's columns values")
    yearInString = 2017
    for row in records:
        year_dict[str(yearInString)] = row[0]
        yearInString += 1
        
       
        

except (Exception, psycopg2.Error) as error:
    print("Error while fetching data from PostgreSQL", error)


cursor.close() 
```
